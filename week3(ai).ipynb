```python
import pandas as pd

# Load Excel dataset
df = pd.read_excel("/content/Food_Recommendation_Dataset.xlsx")

print("=" * 55)
print("        🍽️ AI FOOD RECOMMENDATION SYSTEM 🍽️")
print("=" * 55)

# ---------------- Cuisine Menu ----------------
print("\nChoose Your Cuisine:")
cuisines = df["Cuisine"].unique()

for i, cuisine in enumerate(cuisines, start=1):
    print(f"{i}. {cuisine}")

choice = int(input("\nEnter your choice: "))
cuisine = cuisines[choice - 1]

# ---------------- Food Type ----------------
print("\nChoose Food Type:")
print("1. Veg")
print("2. Non-Veg")

type_choice = int(input("Enter your choice: "))

if type_choice == 1:
    food_type = "Veg"
else:
    food_type = "Non-Veg"

# ---------------- Meal ----------------
print("\nChoose Meal:")
print("1. Breakfast")
print("2. Lunch")
print("3. Dinner")
print("4. Snacks")

meal_choice = int(input("Enter your choice: "))

meal_dict = {
    1: "Breakfast",
    2: "Lunch",
    3: "Dinner",
    4: "Snacks"
}

meal = meal_dict[meal_choice]

# ---------------- Spice Level ----------------
print("\nChoose Spice Level:")
print("1. Mild")
print("2. Medium")
print("3. Spicy")

spice_choice = int(input("Enter your choice: "))

spice_dict = {
    1: "Mild",
    2: "Medium",
    3: "Spicy"
}

spice = spice_dict[spice_choice]

# ---------------- Budget ----------------
budget = int(input("\nEnter Your Budget (₹): "))

# ---------------- Recommendation ----------------
recommend = df[
    (df["Cuisine"] == cuisine) &
    (df["Type"] == food_type) &
    (df["Meal"] == meal) &
    (df["Spice"] == spice) &
    (df["Budget"] <= budget)
]

print("\n" + "=" * 55)
print("🍴 YOUR FOOD RECOMMENDATIONS")
print("=" * 55)

if recommend.empty:
    print("\n😔 Sorry! No food found based on your preferences.")
else:
    recommend = recommend.sort_values(by="Rating", ascending=False)

    for i, (_, row) in enumerate(recommend.iterrows(), start=1):
        print(f"\n🍽️ Recommendation {i}")
        print("-" * 35)
        print(f"Food      : {row['Food']}")
        print(f"Cuisine   : {row['Cuisine']}")
        print(f"Type      : {row['Type']}")
        print(f"Meal      : {row['Meal']}")
        print(f"Spice     : {row['Spice']}")
        print(f"Price     : ₹{row['Budget']}")
        print(f"Rating    : ⭐ {row['Rating']}")

print("\n🙏 Thank you for using AI Food Recommendation System!")
```
